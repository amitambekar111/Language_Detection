```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/TheiScale/YouTube-Video-Notes/refs/heads/main/Machine%20Learning%2BNLP%20Project%20Language%20Detection/language.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['language'].value_counts()
```

```python
df.dtypes
```

```python
x = np.array(df['Text'])
y = np.array(df['language'])
```

```python
print(x)
```

```python
print(y)
```

```python
cv = CountVectorizer()
X = cv.fit_transform(x)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y ,test_size=0.33,random_state=42)
```

```python
X_train
```

```python
print(X_train)
```

```python
y_test
```

```python
model = MultinomialNB()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
user = input("Enter a Text: ")
df = cv.transform([user]).toarray()
output = model.predict(df)
print(output)
```
